## 3. How many Vegetarian and Meatlovers were ordered by each customer?

**Expected Output:**

image.png

```python
# Write Your code here.
mysql('''
SELECT c.customer_id, p.pizza_name, COUNT(p.pizza_name) AS order_count
FROM customer_orders AS c
JOIN pizza_names AS p
 ON c.pizza_id= p.pizza_id
GROUP BY c.customer_id, p.pizza_name
ORDER BY c.customer_id;
''')
```

## 4. What was the maximum number of pizzas delivered in a single order?

**Expected Output:**

image.png

```python
# Write Your code here.
mysql('''
SELECT MAX(pizza_per_order) AS max_no_of_pizzas_in_single_order 
from
(
SELECT c.order_id, COUNT(c.pizza_id) AS pizza_per_order
FROM customer_orders AS c
JOIN runner_orders AS r
ON c.order_id = r.order_id
WHERE r.distance != 0
GROUP BY c.order_id) As T

''')
```
